```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
import os
import requests


chrome_options = Options()
chrome_options.add_argument("--headless")  
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")

# Use Service class for WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

# eBay URLs
urls = [
    "https://www.ikea.com/us/en/rooms/living-room/gallery/",
    "https://www.gettyimages.com/photos/furnished-rooms"
]


os.makedirs("NiceFurnishedRooms", exist_ok=True)

for url_index, url in enumerate(urls, start=1):
    print(f"Scraping images from Furnished Rooms URL {url_index}...")

    
    driver.get(url)
    
    
    time.sleep(5)
    images = driver.find_elements(By.TAG_NAME, "img")

    # Download images
    for i, img in enumerate(images):
        img_url = img.get_attribute("src")
        if img_url and "https" in img_url:
            try:
                img_data = requests.get(img_url).content
                with open(f"NiceFurnishedRooms/image_{url_index}_{i}.png", "wb") as f:
                    f.write(img_data)
                print(f"Downloaded image_{url_index}_{i}.png")
            except Exception as e:
                print(f"Could not download image_{url_index}_{i}: {e}")

# Close
driver.quit()

print("Scraping completed!")
```
